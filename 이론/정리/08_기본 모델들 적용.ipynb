# 3. 데이터 전처리

* 생략

```
song_data = pd.DataFrame(data = song_data)
song_data
```

# 4. 머신러닝을 위한 데이터 전처리

```
# train, test split
# 이러면 song_data 내부에 학습을 결정지을 요소로 song_popularity가 있는 것 아닌가?? -> 제거
from sklearn.model_selection import train_test_split
song_popularity = song_data['song_popularity']
song_data = song_data.drop('song_popularity', axis = 1)
```

```
song_data.info()
```

```
song_data.set_index('song_name', inplace=True)
```

```
X_train, X_test, y_train, y_test = train_test_split(song_data, song_popularity, random_state = 42)
```

```
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

# 5. 학습

## 5.1 선형 모델

```
from sklearn.linear_model import LinearRegression
lin = LinearRegression()
lin.fit(X_train, y_train)
```

## 5.2 SVM

```
from sklearn.svm import SVR
```

```
model = SVR(kernel = 'rbf', gamma = 'scale')
model.fit(X_train, y_train)
```

## 5.3 결정 트리

```
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import export_graphviz
from sklearn.model_selection import cross_val_score
import graphviz

model1 = DecisionTreeRegressor(max_depth = 10)
model1.fit(X_train, y_train)
```

## 5.4 랜덤 포레스트

```
from sklearn.ensemble import RandomForestRegressor
```

```
model2 = RandomForestRegressor()
model2.fit(X_train, y_train)
```

# 6. 평가 및 튜닝
* 선형 모델, SVM, 결정 트리, 랜덤 포레스트의 평가 값을 확인

## 선형 모델

```
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import cross_val_score
```

```
song_predictions = lin.predict(X_test)
print(f"예측 결과 : {song_predictions}")
```

```
linear_result = X_test[:] # 깊은 복사 필요
linear_result['song_predictions'] = song_predictions
```

```
linear_result # 끝에 각 곡에 맞는 예측 결과
```

```
from sklearn.metrics import mean_squared_error

lin_mse = mean_squared_error(song_predictions, y_test)
lin_rmse = np.sqrt(lin_mse)
print(lin_rmse)
```

```
from sklearn.metrics import mean_absolute_error

lin_mae = mean_absolute_error(song_predictions, y_test)
print(lin_mae)
```

## SVM

```
print(f"학습 : {model.score(X_train, y_train)}")
print(f"예측 : {model.score(X_test, y_test)}")
```

## 결정 트리(Decision Tree)

```
cross_val_score(estimator = model1, X = song_data, y = song_popularity, cv = 5)
```

```
dot_data = export_graphviz(decision_tree = model1,
                           feature_names = song_data.columns.values,
                           filled = True,
                           rounded = True)

graph = graphviz.Source(dot_data)
graph
```
