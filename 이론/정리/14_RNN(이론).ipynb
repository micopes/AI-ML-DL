# 순환 신경망

- 순환 신경망(RNN)은 CNN과 입력을 받아들이는 방식과 처리하는 방식에 차이가 있다.
- 순환 신경망은 **순서가 있는 데이터**를 입력으로 받고, 같은 네트워크를 이용해 변화하는 입력에 대한 출력을 얻어낸다
  - 순서가 있는 데이터는 음악, 자연어, 날씨, 주가 등 시간의 흐름에 따라 변화하고 그 변화가 의미를 갖는 데이터

## 순환 신경망의 구조
- RNN은 기존의 신경망과 달리 되먹임(feedback)구조를 가지고 있다
  - 되먹임(feedback)이란 어떤 레이어의 출력을 다시 입력으로 사용하는 것

image.png

### Feedback
- RNN은 출력값이 다음 입력에 사용되고 있기 때문에 RNN 네트워크에 동일하게 전달된다는 점이 가장 큰 특징
- 모든 RNN 레이어에서 입력과 출력이 되먹임 구조를 활용하고 있다는 점을 반드시 이해해야 한다.

## RNN의 특징
- RNN은 **입력과 출력의 길이에 제한이 없다.**
- 이런 특징을 활용하면 **다양한 형태**의 신경망을 구성하는 것이 가능하다.

## LSTM
- SimpleRNN 레이어는 입력 데이터가 길어질수록, 학습 능력이 떨어진다는 점이 매우 큰 단점 -> 장기 의존성 문제(Long-Term Dependency)
- 입력 데이터와 출력 사이의 길이가 멀어질수록 연관 관계가 적어진다 -> 과거 정보에 의존해야 하는 RNN의 특성상 과거 시점이 현재와 너무 멀어지면 문제를 풀기 힘들어진다.
- 해당 문제를 해결하기 위해서 LSTM(Long Short Term Memory)이 제안됨

### LSTM의 핵심 아이디어
- Simple RNN 셀에선 다음 과정의 방향으로 출력이 전달되는데 비해서 LSTM셀은 셀 상태를 나타내는 C와 평행선을 그리며 데이터를 전달 -> 다음 진행 과정을 가로지르며 셀 상태가 보존되기 때문에 장기의존성 문제를 해결할 수 있다는 것이 LSTM의 핵심 아이디어

## GRU 레이어
- LSTM 구조를 더 간단하게 만들어서 계산의 이점을 살린 것이 GRU이며, 어떤 문제는 LSTM보다 좋은 성능을 보기도 한다.
- GRU는 셀 상태가 없으며, 기존의 LSTM보다 Sigmoid 함수를 적게 사용(게이트를 줄임)했다는 것을 의미

## 임베딩(Embedding) 레이어
- 임베딩 레이어는 자연어를 수치화된 정보로 바꾸기 위한 레이어
  - 자연어도 정보를 잘게 분리할 수 있는데, 영어는 문자 단위로 가능하며, 한글은 자소 단위로 가능, 좀 더 큰 단위인 띄어쓰기도 가능
  - 몇 개의 문자를 묶어서 파악하는 n-gram 기법 등도 사용
  - 정수 및 실수를 사용해서 수치화를 진행
- 임베딩 레이어를 학습시키는 방법은 다양
  - Word2Vec, GloVe, FastText, ELMo 등이 존재
