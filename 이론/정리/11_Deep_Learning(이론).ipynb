# 딥러닝 모델 작성 순서

1) 데이터셋 생성하기
- 원본 데이터를 불러오거나 시뮬레이션을 통해 데이터를 생성
- Train/Val/Test 생성
- 딥러닝 모델의 학습 및 평가를 할 수 있도록 포맷 변환

2) 모델 구성하기
- 시퀀스 모델을 생성한 뒤 필요한 **레이어** 구성(*레이어가 어떻게 구성되어 있는가?*)

3) 모델 학습과정 설정하기
- 학습에 대한 설정을 수행
- **손실 함수 및 최적화 방법**을 정의
- compile() 함수를 사용

4) 모델 학습시키기
- 훈련 셋을 이용하여 구성한 모델로 학습
- fit() 함수를 사용

5) 학습 과정 살펴보기
- 모델 학습 시 Train/Val의 Loss 및 정확도를 측정(**반복 횟수** 조정)

6) 모델 평가하기
- 준비된 시험 셋으로 학습한 모델을 평가

7) 모델 사용하기
- predict() 함수를 사용

## 주요 Layer

### 1. Dense(내가 알고있는 일반 뉴런)
- 다층 퍼셉트론 신경망에서 사용되는 레이어
- 입력과 출력을 **모두** 연결해 준다
- 입력 뉴런이 4개, 출력 뉴런을 8개라고 할때 연결 뉴런은 32개이다.

### 2. Convolution Layer
- Convolution 신경망
- 다층 퍼셉트론 신경망과 유사하나 이미지가 가지고 있는 특성이 고려되어 설계된 신경망으로 영상 처리에 주로 사용
- Max Pooling Layer, Flatten Layer

### 3. LSTM(Long Short-Term Memory units)
- 순환 신경망 모델
- 순차적인 자료에서 규칙적인 패턴을 인식하거나 그 의미를 추론
- 순차적이라는 특성 때문에 간단한 레이어로도 다양한 형태의 모델을 구성할 수 있는 곳에 주로 사용

## 손실 함수 및 최적화
손실함수(혹은 목적(Objective) 함수)
- 훈련하는 **동안** 최소화될 값 -> 주어진 문제의 성공 지표
- Binary Crossentropy
  - 이진 분류에서 사용
- Categorical Crossentropy
  - 다항 분류에서 사용
- CTC(Connection Temporal Classification)
  - 시계열

최적화
- 손실 함수를 기반으로 네트워크가 어떻게 업데이트 되는지를 결정
  - SGD
- 학습이 시작되면 blackbox이므로, 최적화 알고리즘을 이용해야 한다.
