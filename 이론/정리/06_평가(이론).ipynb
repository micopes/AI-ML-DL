# (결과)평가

- 좋은 모델을 생성하기 위해 평가를 사용.
- 그렇다면 나쁜 모델은?
  - 과적합
    - 학습이 잘됨
    - 예측 성능 떨어짐(generalization)
    - **해결책** :
      - 1. 데이터를 늘인다
      - 2. 규제(L1, L2), 
      - 3. Dropout
      - 4. 잡음 주입
      - 5. Batch Normalization
  - 과소적합
    - 학습 잘 안됨
    - 예측 성능 떨어짐(generalization)
    - **해결책** : 
      - 1. 모델 복잡도 높인다.
      - 2. 제약 조건을 완화한다.

- 회귀와 분류는 평가 지표가 다르다.
- 결과값의 차이가 분명하기 때문에 회귀와 분류에 따라서 평가 지표를 분리해서 사용해야 한다.

### MAE(Mean Absolute Error)
- 모델의 예측값과 실제값의 차이를 모두 더한다는 개념(총 개수로 나눈다)
- 절댓값을 취하기 때문에 가장 직관적으로 알 수 있는 지표
- **MSE 보다 특이치에 robust**
- 절댓값을 취하기 때문에 모델이 underperformance인지 overperformance인지 **알 수 없음**
  - underperformance : 모델이 실제보다 낮은 값으로 예측
  - overperformance : 모델이 실제보다 높은 값으로 예측

- 제곱을 하기 때문에 MAE와는 다르게 모델의 예측값과 실제값 차이의 면적의 합 => 특이값이 존재하면 수치가 많이 늘어남
- 특이치에 민감(제곱하기 때문에)

### RMSE (Root Mean Squared Error)
- RMSE를 사용하면 오류 지표를 실제 값과 유사한 **단위**로 다시 변환하여 해석을 쉽게 함(주가 분석, 보험금, 투자 금액 등 **단위기반** 오차 분석)
- 가장 많이 사용됨

### MAPE(Mean Absolute Percentage Error)
- MAPE는 MAE를 퍼센트로 변환한 것
- MAE와 마찬가지로 MSE보다 특이치에 robust하며, MAE와 같은 단점을 가짐
- 추가적으로 모델에 대한 **편향**이 존재
  - 이 단점에 대응하기 위해 MPE도 추가로 확인
  - **(오차가)0 근처의 값에서는 사용하기 어려움**

### 기타

- MSLE(Mean Squared Log Error)
  - MSE에 log를 적용해준 지표
  - log(y)가 아니라 log(y+1)을 사용, y = 0일 때, log(y)는 마이너스 무한대이므로 이를 보정해주기 위해서 log(y+1)을 사용

- RMSLE (Root Mean Squared Log Error)
  - RMSE에 log를 적용해준 지표

- R 제곱(R Square)
  - R square는 분산 기반으로 예측 성능을 평가
  - 1에 가까울수록 예측 정확도가 높음
  - R square = 예측값 Variance / 실제값 variance

### 정확도

- 예측 결과가 동일한 데이터 / 전체 예측 데이터
- 가장 직관적으로 사용할 수 있는 지표. 
  - 그러나, 데이터의 label이 불균형할 경우
  성능을 왜곡할 수 있는 문제점을 가지고 있음(ex) 신용카드 사기 검출)

### 정밀도
  - TP / (FP + TP)
  -  이게 업계에서의 정확도 정밀도는 Positive 로 예측한 대상들이 실제로 맞는(True)인 비율
  - 양성 예측도라고 불리기도 함

### 재현율
  - TP / (TP + FN)
  - 재현율은 예측과 실제 값이 positive로 일치한 비율
  - 민감도 혹은 TPR(True Positive Rate)로 불리기도 함
  - 다시 예측해도 크게 변하지 않음(모델이 robust할수록 더.). 편향된 데이터에 유용

이익을 낼때는 정밀도, 무엇을 평가하기 위해서는 재현율.


### F1 Score
  - 정밀도와 재현율의 조화 평균
  - 정밀도와 재현율의 trade-off에서 오는 평가의 문제를 완화
