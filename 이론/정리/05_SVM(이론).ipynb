## 소프트 마진 분류


- 모든 샘플이 도로 바깥쪽에 위치하면 이를 하드 마진 분류 라고 함
  - **데이터가 선형적으로 구분되어야 함**
  - **이상치에 민감함**
- 하드 마진의 단점을 해결하기 위해서
1. 도로의 폭을 가능한 넓게 유지
2. 마진 오류 사이의 적절한 균형을 잡은 것 : 소프트 마진 분류


- 선형적으로 완벽하게 하드마진 하라고 하는 경우 이상치를 없애야 한다.

## 비선형 SVM 분류

image.png

ex) **X -> X^2**

- 선형 분류가 안되는 데이터를 **다항 특성**같은 별도의 특성을 추가하여 비선형 데이터로 만드는 것
- 다항 특성을 추가하는 것은 간단하고 대부분 정상동작하지만 낮은 차수의 특성은 데이터를 잘 표현하지 못할 가능성이 높고, 높은 차수의 특성은 너무 많은 다항특성을 추가해 계산이 매우 복잡해진다.

- Kernel Trick
  - 다항 특성 추가는 간단하고 대부분의 경우에서 잘 작동하지만, 차수가 높아질수록 계산량이 폭증한다는 치명적인 단점, 계산량이 증가하게 되면 그만큼 최적의 초평면을 구하는데 오래 걸리기 때문에 다른 방법을 사용한다.
  - Kernel Trick이라는 수학적 기교를 사용하여 변환 함수의 정보 없이 실제로 다항 특성을 추가하지 않더라도 다항 특성을 여러 개 추가한 것과 같은 결과를 얻을 수 있다.
  - SVM을 사용하면 **커널 트릭** 이라는 수학적 기교를 적용할 수 있다.

속도 : 선형 > 커널 >> 다항식
