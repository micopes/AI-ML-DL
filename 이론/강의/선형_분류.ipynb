# 분류(Classification)

모델 기반 학습(모델 - 수식)
- 모델이 최상의 성능을 내도록 하기 위해서
  - 모델이 얼마나 좋은지 측정하는 **효용(Utility)** 함수
  - 모델이 얼마나 나쁜지 측정하는 **비용(Cost)** 함수
를 정의할 수 있다.

베이즈 정리 : 
- 암인데 암이아니라고 할 확률
- 암이 아닌데 암이라고 할 확률
..

## 로지스틱 회귀(분류에 사용)
- 로지스틱 회귀는 선형 함수를 통과시켜 얻은 z를 임계함수에 보내기 전에 변형시키는데 이를 활성화 함수(Activation Function)이라고 한다.
- 선형 -> 활성화 -> 임계

퍼셉트론 
- **이진 분류** 문제에서 사용되었었음.
- 임의의 샘플 데이터를 True나 False로 구분하는 문제

- 활성화 함수 : 값을 0과 1사이 값으로 만들어줘야 함.
- 임계함수 : 임계값에 따라 0, 1로 나뉨(ex) 0.5)

활성화 함수는 비선형 함수
- 활성화 함수가 선형 함수가 되면 값이 계속 커지게 된다. 그렇게 되지 않고 0과 1로 수렴되게 하기 위해서 비선형 함수를 사용한다.
